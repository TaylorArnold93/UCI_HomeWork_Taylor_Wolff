```python
%matplotlib inline
from sqlalchemy import create_engine
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
engine = create_engine(f'postgresql://postgres:Family34@localhost:5432/EmployeeSQL_DB')
connection = engine.connect()
```

```python
Employee_df=pd.read_sql('Select * From Employees', connection)
```

```python
Salaries_df=pd.read_sql('Select * From Salaries', connection)

Salaries_df.head()
```

```python
salaries=[]

for x in np.arange(len(Salaries_df['salary'].values)):
    response=Salaries_df['salary'][x]
    salaries.append(response)
    x+=1
```

```python
# Creating Histogram
plt.hist(salaries)
plt.xlabel('Salary')
plt.ylabel('Employee Count')
plt.title()
plt.show()
```
